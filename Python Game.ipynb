```python
from tkinter import *
from PIL import Image,ImageTk
from random import randint

#main windows
root = Tk()
root.title("Piepdra, Papel o Tijeraaaaaa!!!")
root.configure(background="#c7d8f9")


# Imagen
paper_img_PC = ImageTk.PhotoImage(Image.open("Papel PC.png"))
rock_img_PC = ImageTk.PhotoImage(Image.open("Piedra PC.png"))
Scissor_img_PC = ImageTk.PhotoImage(Image.open("Tijera PC.png"))

paper_img = ImageTk.PhotoImage(Image.open("Papel User.png"))
rock_img = ImageTk.PhotoImage(Image.open("Piedra User.png"))
Scissor_img = ImageTk.PhotoImage(Image.open("Tijera User.png"))

# Insertar imagenes
user_label = Label(root, image= paper_img, bg="#c7d8f9")
comp_label = Label(root, image= rock_img_PC, bg = "#c7d8f9")

comp_label.grid(row = 1, column = 4)
user_label.grid(row = 1, column = 0)

#Puntaje
playerScore = Label(root, text = 0, font = 100, bg ="#c7d8f9", fg = "#221c21")
PcScore = Label(root, text = 0, font = 100, bg ="#c7d8f9", fg = "#221c21")

playerScore.grid(row = 1, column = 1)
PcScore.grid(row = 1, column = 3)

#Indicadores
user_indicator = Label(root, font =50, text ="User",bg ="#c7d8f9", fg = "#221c21")
PC_indicator = Label(root, font =50, text ="Computadora", bg ="#c7d8f9", fg = "#221c21")

user_indicator.grid(row=0, column=1)
PC_indicator.grid(row=0, column=3)


#mensajes
msg = Label(root, font= 50, bg="#9b59b6", fg="#221c21")
msg.grid(row=3, column=2)


#Update Message
def updateMessage(x):
    msg["text"] = x

#Update User Score
def updateUserScore():
    score = int(playerScore["text"])
    score += 1
    playerScore["text"] = str(score)
    
#Update PC Score
def updatePcScore():
    score = int(PcScore["text"])
    score += 1
    PcScore["text"] = str(score)
    

#Check Winner
def checkWin(player, Computer):
    if player == Computer:
        updateMessage("Es un empate!")
    elif player == "rock":
        if Computer == "paper":
            updateMessage("Perdiste...")
            updatePcScore()
        else:
            updateMessage("Ganaste!")
            updateUserScore()
            
    elif player == "paper":
        if Computer == "scissor":
            updateMessage("Perdiste...")
            updatePcScore()
        else:
            updateMessage("Ganaste!")
            updateUserScore()
            
    elif player == "scissor":
        if Computer == "rock":
            updateMessage("Perdiste...")
            updatePcScore()
        else:
            updateMessage("Ganaste!")
            updateUserScore()
    else:
        pass
    

#actualizar opciones
choices = ["rock","paper", "scissor"]
def updateChoice(x):
    
# Opciones PC
        compChoice = choices[randint(0,2)]
        if compChoice == "rock":
            comp_label.configure(image = rock_img_PC)
        elif compChoice == "paper":
            comp_label.configure(image = paper_img_PC)
        else:
            comp_label.configure(image = Scissor_img_PC)


        # Opciones Usuario
        if x=="rock":
            user_label.configure(image= rock_img)
        elif x=="paper":
            user_label.configure(image= paper_img)
        else:
            user_label.configure(image= Scissor_img)
        
        checkWin(x,compChoice)


# Botones
piedra = Button(root, width = 20, height = 2, 
                text = "Piedra", bg ="red",command = lambda:updateChoice("rock"), fg ="white" ).grid(row=2, column=1)
papel = Button(root, width = 20, height = 2, 
               text = "Papel", bg ="blue",command = lambda:updateChoice("paper"), fg ="white" ).grid(row=2, column=2)
tijera = Button(root, width = 20, height = 2, 
                text = "Tijera", bg ="green",command = lambda:updateChoice("scissor"), fg ="white" ).grid(row=2, column=3)



root.mainloop()
```
